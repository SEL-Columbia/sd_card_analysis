```python
# read in csv to data frame
import pandas as pd
import matplotlib.pyplot as plt
sd_watts = pd.read_csv('/Users/dsoto/repos/processed_ss_data/sd_card/sd_watts_5min.csv', 
                        index_col='date', 
                        parse_dates=True)
```

```python
sd = '2012-01-01'
ed = '2012-02-01'

sd_watts = sd_watts[sd:ed]

import numpy as np
bins = np.linspace(0, 100, 101)

# create M by N array where M is circuit and N is the histogram
mylist = []
labels = []
for col in sd_watts.columns:
    if '_00' in col:
        continue
    if 'ug' in col:
        continue
    '''
    if 'ml08_16' in col:
        continue
    if 'ml06_16' in col:
        continue
    '''
    
    # trim off low wattage
    data = sd_watts[col].dropna()
    #data[data<0] = 0
    
    values, bins = np.histogram(data, bins=bins)
    # energy based feature vector
    mylist.append(values * bins[:-1])
    
    # boolean feature vector
    #values[values > 1] = 1
    #mylist.append(values)
    
    # create list of columns
    labels.append(col)
    
data = np.array(mylist)
#data = data[:, 10:]
#bins = bins[10:]
# run clustering

# trying two approaches
# one, basically is a boolean vector of whether or not power was consumed at that level
# this predicted some of the ones i picked visually
# two, having the bin be the energy consumed at each power level
# this seems promising, but i'm having matrix issues

# i'm having problems with the matrix not being positive definite

'''
import scipy.cluster.vq
a, b = scipy.cluster.vq.kmeans2(data, 2)
'''
k = 3
from sklearn import cluster
kmeans = cluster.KMeans(k=k)
kmeans.fit(data)
b = kmeans.labels_


# output table of circuit labels and cluster results

for i in range(k):
    plt.plot(bins[:-1],kmeans.cluster_centers_[i])
    plt.title(str(i) + ' profile')
    plt.show()

label_list = []

categories = pd.Series(data=b, index=labels)
for i in range(k):
    label_list.append(categories[categories==i])
    print i, len(label_list[i])
    
    '''
    one = categories[categories==1]
zero = categories[categories==0]

print 'one length =', len(one)
print 'zero length =', len(zero)'''
```
