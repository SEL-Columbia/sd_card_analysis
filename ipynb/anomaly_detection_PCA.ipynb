```python
# read in csv to data frame
%pylab inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
sd_watts = pd.read_csv('/Users/dsoto/repos/processed_ss_data/sd_card/sd_watts_5min.csv', 
                        index_col='date', 
                        parse_dates=True)
```

```python
col = 19
ds = '2012-01-01'
de = '2012-06-01'

# for a given circuit, we create a list of vectors with each day of timeseries data

col_name = sd_watts.columns[col]
print col_name

def svd_for_col(col_name):
    
    # filter circuits with too much missing data
    data = sd_watts[col_name][ds:de].fillna(method='pad')
    #data = sd_watts[sd_watts.columns[col]][ds:de].fillna(0)
    
    
    # spit out values and reshape into numpy array
    
    da = np.reshape(data.values, (153, 288))
    
    cov = np.cov(da, rowvar=0)
    
    try:
        (U, S, V) = np.linalg.svd(cov)
        import ex7
        projected = ex7.project_data(da, U, 2)
        scatter(projected[:,0], projected[:,1])
        plt.show()
    except LinAlgError:
        print 'didnt converge'
    
for c in range(len(sd_watts.columns)):
    svd_for_col(sd_watts.columns[c])
```
