# Principal Component Analysis on timeseries data

This script finds an aggregated time series for each circuit in the data set and then projects them onto a subspace using PCA.

```python
# read in csv to data frame
%pylab inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


sd_watts = pd.read_csv('/Users/dsoto/repos/processed_ss_data/sd_card/sd_watts_5min.csv', 
                        index_col='date', 
                        parse_dates=True)
```

## Assemble mean or median timeseries in a dataframe

```python
#col = 'ml05_11'
ds = '2012-01-01'
de = '2012-06-01'

# algorithm to average timeseries

data_list = []
labels = []

mean_power = {}
median_power = {}
for col in sd_watts.columns:
#for col in ['ml05_11', 'ml05_12']:
    if '_00' in col:
        continue
    # skip uganda meters
    if 'ug' in col:
        continue
    # loop over days in range
    # subselect one circuit of data over time frame and replace missing with zeros
    data = sd_watts[col][ds:de].fillna(0)
    # 288 samples per day for 5 minute data
    stride = 288
    # accumulator 
    a = np.zeros(stride)
    for i in range(0, 8640, stride):
        a = np.vstack((a, data.ix[i:i + stride]))
    # could use median instead of mean here
    median_power[col] = np.median(a, 0)
    mean_power[col] = np.mean(a, 0)
    
mean_power = pd.DataFrame(mean_power)
median_power = pd.DataFrame(median_power)
```

## Calculate the covariance matrix and SVD

```python
import numpy as np
cov = np.cov(median_power)
#cov = np.cov(mean_power)
(U, S, V) = np.linalg.svd(cov)
```

## Plot the N-th eigenvector

These should represent the vectors of highest variance in the data set.
